```python
# Dependencies
from splinter import Browser
import pandas as pd
from itertools import cycle
import numpy as np
```

```python
# Pulling data from this url
url = 'https://www.espn.com/nfl/fpi/_/view/efficiencies/season/2019'
```

```python
# Reading the HTML tables into a list of tables 
tables = pd.read_html(url, header=0)
```

```python
df = tables[0]
```

```python
df.index = np.arange(1, len(df)+1)
df
```

```python
df1 = tables[1]
```

```python
df1 = df1.rename(columns=df1.iloc[0]).drop(df1.index[0])
df1
```

```python
eff_19 = df.join(df1, how='outer')
eff_19
```

```python
eff = cycle(['2019'])
eff_19['Year'] = [next(eff) for n in range(len(eff_19))]
```

```python
eff_19
```

```python
eff_19.to_csv('results/eff_19.csv',index=False)
```
