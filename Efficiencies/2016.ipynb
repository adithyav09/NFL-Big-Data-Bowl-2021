```python
# Dependencies
from splinter import Browser
import pandas as pd
from itertools import cycle
```

```python
# Pulling data from this url
url = 'https://www.espn.com/nfl/fpi/_/view/efficiencies/season/2016'
```

```python
# Reading the HTML tables into a list of tables 
tables = pd.read_html(url, header=0)
```

```python
df = tables[0]
```

```python
df1 = tables[1]
```

```python
eff_16 = df.join(df1, how='outer')
eff_16
```

```python
eff = cycle(['2016'])
eff_16['Year'] = [next(eff) for n in range(len(eff_16))]
```

```python
eff_16
```

```python
eff_16.to_csv('results/eff_16.csv',index=False)
```
