```python
# Dependencies
from splinter import Browser
import pandas as pd
from itertools import cycle
```

```python
# Pulling data from this url
url = 'https://www.espn.com/nfl/fpi/_/view/efficiencies/season/2017'
```

```python
# Reading the HTML tables into a list of tables 
tables = pd.read_html(url, header=0)
```

```python
df = tables[0]
```

```python
df1 = tables[1]
```

```python
eff_17 = df.join(df1, how='outer')
eff_17
```

```python
eff = cycle(['2017'])
eff_17['Year'] = [next(eff) for n in range(len(eff_17))]
```

```python
eff_17
```

```python
eff_17.to_csv('results/eff_17.csv',index=False)
```
