```python
def cleanWins(record):
    if("." in str(record)):
        return record
    elif (type(record) != float) and ("-" in record):
        regx = re.compile("\d+-")
        wins = regx.search(record)
        if wins != None:
            wins = wins[0][:-1]
        #losses = regx.search(record)[1]
            return wins
```

```python
def cleanLs(record):
    if("." in str(record)):
        record = float(record)
        record = 16 - record
        return record
    elif (type(record) != float) and ("-" in record):
        regx = re.compile("-\d+")
        wins = regx.search(record)
        if wins != None:
            wins = wins[0][1:]
        #losses = regx.search(record)[1]
            return wins
```

```python
df = df.assign(PROJ_WINS2 = lambda x: x["PROJ WINS"].apply(cleanWins))
```

```python
def realWins(name):
    regex = re.compile(".\d+-\d+")
    record = regex.search(name)
    if record != None:
        record = record[0]
        record = record[1:]
        rec = record.split("-")
        wins = rec[0]
        losses = rec[1]
        return wins
    else:
        return 0
```

```python
df = df.assign(REAL_WINS = lambda x: x.TEAM.apply(realWins))
```

```python
def realLss(name):
    regex = re.compile(".\d+-\d+")
    record = regex.search(name)
    if record != None:
        record = record[0]
        record = record[1:]
        rec = record.split("-")
        wins = rec[0]
        losses = rec[1]
        return losses
    else:
        return 0
```

```python
df = df.assign(REAL_LOSSES = lambda x: x.TEAM.apply(realLss))
```

```python
def cleanName(name):
    regex = re.compile("\d+-\d+")
    record = regex.search(name)
    if record != None:
        record = record[0]
        name = name.replace(record, "")
        return name
    else:
        return name
```

```python
df = df.assign(CLEAN_NAME = lambda x: x.TEAM.apply(cleanName))
```

```python
def cleanPercent(s):
    s = str(s)
    s = s.strip()
    if s.startswith(">99"):
        s = "99.99"
    elif s.startswith("<1"):
        s = "1"
    if (">" not in s) and ("%" in s):
        num = s.replace("%", "")
        num = float(num) / 100
        return num
    elif ">" in s:
        num = s.replace(">", "")
        num = s.replace("%", "")
        num = float(num) / 100
        return num
    elif "<" in s:
        num = s.replace("<", "")
        num = float(num) / 100
        return num
    else:
        return 0.003
```

```python
def cleanPoints(pts):
    pts = str(pts)
    if "+" in pts:
        pts = pts.replace("+", "")
        pts = float(pts)
        return pts
    elif "-" in pts:
        pts = pts.replace("-", "")
        pts = float(pts)
        return -pts
    else:
        return 0.00
```
