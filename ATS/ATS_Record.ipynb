```python
# Import Dependencies
from splinter import Browser
import pandas as pd
```

```python
# Read 2020 ATS
url = 'https://www.teamrankings.com/nfl/trends/ats_trends/?range=yearly_2020'

tables = pd.read_html(url, header=0)

df = tables[0]

df.to_csv('ATS/ats_2020.csv', index = False)
```
